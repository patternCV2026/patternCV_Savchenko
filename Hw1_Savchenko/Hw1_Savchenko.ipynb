## Задача 1
### Вкладник поклав 50 000 умовних одиниць на три різних рахунки в три різних банки.
### По першому рахунку виплати становитимуть 5% річних.
### По другому - 7% річних.
### По третьому - 6% річних.
### Відомо, що через рік вкладник отримав:
### З першого та другого банку - 2250 у.е.
### З першого та третього банку - 1400 у.е.
### Скільки умовних одиниць він поклав на кожен рахунок спочатку?

```python
import numpy as np

# x — 1-й банк
# y — 2-й банк
# z — 3-й банк
# x + y + z = 50000
# 0.05x + 0.07y = 2250
# 0.05x + 0.06z = 1400

A = np.array([
    [1, 1, 1],
    [0.05, 0.07, 0],
    [0.05, 0, 0.06]
])

B = np.array([50000, 2250, 1400])

solution = np.linalg.solve(A, B)

print("Розв'язок:")
print(f"1-й банк: {solution[0]:.2f}")
print(f"2-й банк: {solution[1]:.2f}")
print(f"3-й банк: {solution[2]:.2f}")
```

## Задача 2
### На складі зберігаються мобільні телефони iPhone 6, iPhone 11, iPhone 12 моделей.
### Всього 1328 екземплярів.
### Моделей iPhone 6 на 120 менше, ніж моделей iPhone 11.
### Моделей iPhone 6 на 100 більше, ніж моделей iPhone 12.
### Скільки моделей кожного виду на складі?

```python
import numpy as np

# x — iPhone 6
# y — iPhone 11
# z — iPhone 12
# x + y + z = 1328
# x = y - 120  → x - y = -120
# x = z + 100  → x - z = 100

A = np.array([
    [1, 1, 1],
    [1, -1, 0],
    [1, 0, -1]
])

B = np.array([1328, -120, 100])

solution = np.linalg.solve(A, B)

print("Кількість моделей:")
print(f"iPhone 6: {int(solution[0])}")
print(f"iPhone 11: {int(solution[1])}")
print(f"iPhone 12: {int(solution[2])}")
```

## Задача 3
### Складіть канонічне рівняння еліпсоїда, що проходить через три точки A, B, C:
### $$ A(\sqrt3; 0; \sqrt3), B(\sqrt6; \frac{1}{2}; 0), C(1; \frac{1}{\sqrt3}; 1) $$
### Канонічне рівняння еліпсоїда в прямокутній декартовій системі координат має вид:
### $$ \frac{x^2}{a^2} + \frac{y^2}{b^2} + \frac{z^2}{c^2} = 1 $$ 
### Задача полягає у визначенні параметрів:
### $$a^2, b^2, c^2 $$

```python
import numpy as np

A_matrix = np.array([
    [3, 0, 3],
    [6, 0.25, 0],
    [1, 1/3, 1]
])

B_vector = np.array([1, 1, 1])

solution = np.linalg.solve(A_matrix, B_vector)

A_coef, B_coef, C_coef = solution

a2 = 1 / A_coef
b2 = 1 / B_coef
c2 = 1 / C_coef

print("Параметри еліпсоїда:")
print(f"a^2 = {a2:.3f}")
print(f"b^2 = {b2:.3f}")
print(f"c^2 = {c2:.3f}")

print("\nРівняння:")
print(f"x^2/{a2:.3f} + y^2/{b2:.3f} + z^2/{c2:.3f} = 1")
```

## Задача 4
### Складіть рівняння параболи, що проходить через три задані точки:
### (1, 12)
### (3, 54)
### (-1, 2)
### Необхідно знайти коефіцієнти рівняння:
### $$ y = a * x^2 + b * + c $$

```python
import numpy as np

A = np.array([
    [1**2, 1, 1],
    [3**2, 3, 1],
    [(-1)**2, -1, 1]
])

B = np.array([12, 54, 2])

a, b, c = np.linalg.solve(A, B)

print("Коефіцієнти:")
print(f"a = {a}")
print(f"b = {b}")
print(f"c = {c}")

print(f"\nРівняння: y = {a}x^2 + {b}x + {c}")
```

## Задача 5 
### Створіть функцію get_polynom, яка приймає набір точок виду: $$ (x_1, y_1), (x_2, y_2), ..., (x_{n+1}, y_{n+1}) $$ і знаходить коефіцієнти многочлена n-го ступеня.
### python
### import numpy as np
### def get_polynom(coords):
###     ...
###     return np.linalg.solve(a, b)

```python
import numpy as np

def get_polynom(coords):
    n = len(coords) - 1
    
    A = []
    B = []
    
    for x, y in coords:
        row = [x**i for i in range(n, -1, -1)]
        A.append(row)
        B.append(y)
    
    A = np.array(A)
    B = np.array(B)
    
    return np.linalg.solve(A, B)

coords = [(0,1), (1,3), (2,7)]
coeffs = get_polynom(coords)

print("Коефіцієнти многочлена:")
print(coeffs)
```
